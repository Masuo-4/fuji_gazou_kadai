```python
import os
from PIL import Image
# from PIL import Resampling
import cv2
from natsort import natsorted
img_name = "sample_submission.png"

files = os.listdir("テスト用")
files = natsorted(files)
lst = []
for filename in files:
    img = cv2.imread("テスト用/" + filename)
    # print(img.shape)
    
    lst.append(img)
im_h = cv2.hconcat(lst)
# cv2.imwrite(img_name, im_h)

# 画像の読み込み
lst = []
img = im_h
h, w = img.shape[:2]
split_x = 14
split_y = 1
# 画像の分割処理
cx = 0
cy = 0
for j in range(split_x):
    for i in range(split_y):
        split_pic = img[cy:cy+int(h/split_y), cx:cx+int(w/split_x), :]
        lst.append(split_pic)
        # cv2.imwrite(str(j)+'.png', split_pic)
        cy = cy+int(h/split_y)
    cy = 0
    cx = cx+int(w/split_x)
# 結合
im_v = cv2.vconcat(lst)
print(im_v.shape)
cv2.imwrite(img_name, im_v)
```

```python
#横連結画像作成
import os
from PIL import Image
# from PIL import Resampling
import cv2
from natsort import natsorted
import math

def get_concat_v(im1, im2):
    dst = Image.new('RGB', (im1.width, im1.height + im2.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (0, im1.height))
    return dst

def get_concat_h(im1, im2):
    dst = Image.new('RGB', (im1.width + im2.width, im1.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (im1.width, 0))
    return dst

def get_concat_v_multi_blank(im_list):
    _im = im_list.pop(0)
    for im in im_list:
        _im = get_concat_v(_im, im)
    return _im

def get_concat_h_multi_blank(im_list):
    _im = im_list.pop(0)
    for im in im_list:
        _im = get_concat_h(_im, im)
    return _im

files = os.listdir("テスト用")
files = natsorted(files)
lst = []
for filename in files:
    img = Image.open("テスト用/" + filename)
    # print(img.shape)
    lst.append(img)

# 横連結画像作成
im_h = get_concat_h_multi_blank(lst)
im_h.save("横連結.jpg")
im_h.save("横連結.png")

```

```python
# 分割画像作成と、縦に結合
def ImgSplit(im):
    # 読み込んだ画像を200*200のサイズで54枚に分割する
    height = 256
    width = 256*18

    buff = []
    # 縦の分割枚数
    for h1 in range(1):
        # 横の分割枚数
        for w1 in range(14):
            # print(w1 * width, h1 * height,
            # (w1 + 1) * width, (h1 + 1) * height)
            c = im.crop((w1 * width, h1 * height,
                        (w1 + 1) * width, (h1 + 1)* height))
            buff.append(c)
    return buff


if __name__ == '__main__':
    # 画像分割
    lst = []
    # im_hは横連結画像
    im = im_h
    print(im.size[0], im.size[1])
    for  i,ig in enumerate(ImgSplit(im)):
        lst.append(ig)
        # 分割画像保存
        ig.save(f"{i}.png", "PNG")
    print(len(lst))
    
    # 画像縦に結合
    img = get_concat_v_multi_blank(lst)
    img.save("sample_submission.png", "PNG")
    img.save("sample_submission.jpg", "JPEG")
```

```python
# 分割ディレクトリにある画像を縦結合
files = os.listdir("分割")
files = natsorted(files)
lst = []
for filename in files:
    img = Image.open("分割/" + filename)
    # print(img.shape)

    lst.append(img)
im_v = get_concat_v_multi_blank(lst)
im_v.save("縦結合.jpg","JPEG")
im_v.save("縦結合.png", "PNG")
```
